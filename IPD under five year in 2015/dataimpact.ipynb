```python
#CHECK IF DATA CONTAIN MISSING VALUES
null_data = data[data.isnull().any(axis=1)]
print(null_data)
```

# THE FIRST TEN DISEASES WITH LARGE NUMBER OF INCIDENCE

```python
Disease = []
Number_of_Incidence = []
for column in data.columns:
    if column == "Council":
        pass 
    else:
        Disease.append(column)
        Number_of_Incidence.append(data[column].sum())

result = pd.DataFrame({'Disease':Disease, 'Number_of_Incidence':Number_of_Incidence})
top_ten = result.sort_values('Number_of_Incidence',ascending=False)[0:10]
top_ten
```

```python
fig = plt.figure(figsize=(12,8))
sns.barplot(top_ten.Disease,top_ten.Number_of_Incidence,color='#0000CD')
plt.title('TOP 10 DISEASES IN 2015 UNDER IPD UNDER 5 YEARS')
plt.xlabel('Name of Diseases',fontsize=18)
plt.xticks(rotation='vertical')
plt.ylabel('total number of inpatient (IPD)',fontsize=18)
plt.show()
```

```python
#save the result in jpg format
fig.savefig("figures/2015_top_ten_disease.jpg", bbox_inches="tight", dpi=1200)
```

# THE  FIVE REGIONS WITH LARGE NUMBER OF INCIDENCE

```python
data['mean_values']= data.mean(axis=1)
```

```python
data['mean_values'].sort_values(ascending=False)[0:5]
```

```python
top_five = data[['Council','mean_values']].sort_values('mean_values',ascending=False)[0:5]
top_five
```

```python
fig = plt.figure(figsize=(12,8))
sns.barplot(top_five.Council,top_five.mean_values,color='#FF4500')
plt.title('TOP 5 REGION WITH LARGE NUMBER OF INCIDENCE IN 2015 UNDER 5 YEARS')
plt.xlabel('Name of Region',fontsize=18)
plt.xticks(rotation='vertical')
plt.ylabel('total number of inpatient (IPD)',fontsize=18)
plt.show()
```

```python
#save the result in jpg format
fig.savefig("figures/2015_top_five_regions.jpg", bbox_inches="tight", dpi=1200)
```
